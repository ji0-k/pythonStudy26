```python
# 만약 입력값이 몇개가 될지 모를때는?
def add_many(*args) : # *을 붙이면 몇개의 값이 들어오든 # 숫자,인수 args. 주로사용 , 다른 글자도 가능하긴함
    # *의 뜻은 와일드 문자로 김씨로 시작한다. ~ "김*"
    # "*원" ~ 원으로 끝나는 문자
    # "*딸기*" "딸기"라는 문자가 포함된
    result = 0 #결과를 보관하는 변수 ,,숫자로 받아버림 / result = "" 문자로 받아버림
    for i in args : #args 는 튜플로 들어옴
        result = result + i   # 튜플로 되어있는 값을 번갈아 더함
    return result
```

```python
#add_many 테스트 # 다시 체크 !!!!!!!!!!!!!!
result = add_many(1,2,3,6)
print(result)
```
