```python
# 문자열 포매팅
# 가져다 쓸 부분에 문자열 안에 쓰고 %d(숫자일경우) , %s(문자일경우) , 보다는 fprint를 많이 사용
# 여러개 쓸 경우 괄호안에 항목을 컴마로 구분해주면 됨 (a,b,c ...)
# 많이 쓰이지 않고, 콘솔창에 쓰이는거기 때문에 , 보통 print를 사용 하여 출력
today = "20260105+02-05cloudy"

todayis = ""
htemp = ""
Ltmep = ""

year = today[:4]
month = today[4:6]
day = today[6:8]
high = today[8:11]
low = today[11:14]
weat = today[14:]

todayis ="오늘은 %s 년 %s월 %s일 이며, 오늘의 기온은 최고기온 %s도 최저기온 %s도 입니다 " % (year, month, day, high, low)
print(todayis)
```

```python
#문자열 포맷 코드
# %s	문자열(String) %s로하면 자동타입 변환할 수 있다. 출력용이라서,,
# %c	문자 1개(character) '한 글자' ex. 학점 ABCD... 메모리관리에 유리(최대한 가볍게)
# %d	정수(Integer)
# %f	부동소수(floating-point) (소수점)
# %o	8진수 (입력시 에는 0o)
# %x	16진수 (입력시에는 0x)
# %%	Literal % (문자 % 자체로 사용할때 두번 써주기)
```

```python
# 2개이상의 값 넣을때
# 문장안에서 {0},{1},{2},{...} 쓰고 순서대로 .format(A,B,C,D...) 써준다
today = "20260105+02-05cloudy"

todayis = ""
htemp = ""
Ltmep = ""

year = today[:4]
month = today[4:6]
day = today[6:8]
high = today[8:11]
low = today[11:14]
weat = today[14:]

todayis ="오늘은 {0} 년 {1}월 {2}일 이며, 오늘의 기온은 최고기온 {3}도 최저기온 {4}도 입니다 " .format (year, month, day, high, low)
print(todayis)
```
