```python
# 튜플 () 괄호로 둘러 싼다 #리스트 [] 대괄호
# 1개의 값만 넣읐 때는 요소뒤에 쉼표(,) t=(1,) 반드시 붙이기, 소괄호 생략하더라도 튜플처리된다  t = 1,2,3,4
# 요소값의 변화 X , CRUD안됨 수정 및 삭제 X
# 요소값의 고정, 값이 바뀌면 안되는 것들에 사용한다
```

```python
# 튜플 생성은 변수에 넣은 값을 활용하는 용도
t1 = (1, 2, 'a', 'b')
print(t1)
```

```python
type(t1)
```

```python
# 값을 수정해보자 오류 발생 할 것이다.
t1[3] = 8 #튜플 만들때만 소괄호()이고, 인덱스 찾을때는 대괄호[]
# t1에서 3번째 인덱스 'b'를 8로 수정 시도
```

```python
# 값을 삭제해보자 아마 오류
del t1[2]
```
