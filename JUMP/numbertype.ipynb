```python
#숫자형(NumberType) : 숫자로 이루어진 자료형으로
# 123과 같은 정수, 12.34과 같은 실수, 8진법, 16진법 등이다.

#정ㅇ수형(integer) : 양의 정수 , 음의 정수 , 0
a = 123
print(a) #양의정수
```

```python
b = -123
print(b) #음의정수
```

```python
#실수형(flooating-point) : 소수점이 포함된 숫자
c = 1.2
print(c) # 양의 실수
```

```python
d = -3.45
print(d) #음의 실수
```

```python
#지수형식 (과학기술용) 엄청 큰 실수 들
e = 4.24E10 # 42,400,000,000 424억
print(e)
```

```python
f = 4.24E-10
print(f)
```

```python
#8진수 입력기법 (octal)
#숫자 0과 영문자 O를 쓰고 입력한다.
g = 0o177 #8진법 177
print (g)
```

```python
#16진법 (hexadecimal) 0x로 입력
h=0x8ff #16진법 8ff
print(h)
```

```python
#사칙연산
# a + b 더하기
# a - b 빼기
# a * b 곱하기
# a ** b 제곱
# a / b 나누기
# a//b 몫 구하기
# a % b 나머지 구하기
```

```python
#복합 연산자
#산술연산자와 대입연산자를 합쳐 놓음
i = 0
i = i + 1
print(i)
```

```python
i += 1 # [ i = i + 1 ] 을 줄인 형태
print(i)
```

```python
i -= 1 # [ i = i - 1 ] 을 줄인 형태
print(i)
```

```python
i *= 5 # [ i = i * 5 ] 를 줄인 형태
print(i)
```

```python
# +=, -=, *=, /=, //=, %=, **= 와 같은 것들이 있다
```
