```python
#파일로 더미데이터 및 추가자료를 관리해본다.
# 지금까지는 변수선언할때 초기값을 리스트에 담아 제공 받았다
# 그러나 프로그램 종료하면 추가한 값은 사라진다
# 자료를 영구적으로 보관하는 방법은 파일로 저장하던지 데이터베이스에 보관

# 서버 : 24시간 , 365일 동안 멈춤없이 서비스를 제공하는 컴퓨터 , 서비스 제공자
#
# 클라이언트 : 서비스제공받는자

# 파일 저장 : .txt, csv(엑셀), json(웹api) 등으로 저장
```

```python
f = open("새파일.txt", "w" ) #프로젝트목록에서 디스크에서 다시로드 누르면 새파일txt뜬다, 불렀는데 아무것도없어서 자기가 만들었음
# open은 파일을 열때 사용
# 옵션 : r-> read 읽기전용
#       w-> write 쓰기가능
#       a -> append 추가가능(마지막행에 추가)
# 파일을 열때 경로를 지정하지 않으면 같은 위치에 파일을 찾거나 쓴다.

f.close()
# 파일은 윈도우 정책에서 단일처리를 한다. (열고 작업이 끝나면 닫고 나와야 함)
# 안 닫으면 다른프로그램에서 열 수 없음.
# 모든 파일은 열고 닫기 까지 해주고 나오기
```

```python
f = open("C:/MBC/새파일.txt", "w")
f.close()
# 주의사항 경로에 / 를 붙이는게 일반적이고
# 만약 \를 사용하고 싶으면 \\두개를 사용해야한다.
# \t 탭, \n 엔터 \는 이스케이프 문자로 기능이 있는 문자열
```

```python
f = open("C:/MBC/새파일.txt", "w")
for i in range(1,11) : # range = 1부터 10까지의 값을
    data = "%d번째 줄입니다. \n"  %i
    #         정수 처리    엔터  for i
    f.write(data) # .write () 기록 ! 파일에 저장한다는 뜻
f.close() # 닫고 나오기
```

```python
#파일에 있는 내용을 가져와 보자
#.readline() 파일에 있는 내용 줄단위로 가져오기 (행)
f = open("C:/MBC/새파일.txt", "r") # r 은 읽기 전용모드
line = f.readline() # 기본값은 첫줄 가져오기
print(line) #
f.close()
```

```python
# 모든 줄을 읽어 화면에 출력하고 싶으면
f = open("C:/MBC/새파일.txt", "r")
while True : # False 처리 될 때 까지 반복문
    line = f.readline() # 1줄씩 읽는다
    if not line :
        break
    else :
        print(line)
f.close()

# 메모장 내용에 엔터되어있는데?
# \n\r 라인피드 캐리지 리턴 안보이는 상태로 입력되있음

```

```python
# .readlines() a메모장 파일을 리스트로 한번에 불러온다.
#  파일의 모든 줄을 읽어 리스트 형태로 반환
# 반환값은 문자열 리스트 ['첫번째줄\n','두번째줄\n', ...]
# 파일 내용을 모두 메모리에 로드하기 때문에 파일크기가 클경우 메모리 부족!
# 장점 : 파일의 모든 줄을 한번에 가져와 처리해야 할 때 편리

f =open("c:/MBC/새파일.txt", "r")
lines = f. readlines() # 파일에 있는 모든 내용을 리스트 처리
print(lines)
f.close()
```

```python
f =open("c:/MBC/새파일.txt", "r")
lines = f. readlines() # 파일에 있는 모든 내용을 리스트 처리
#print(lines)
for line in lines :
    print(line)

f.close()
```

```python
# 메모장에 엔터효과인 \n을 지우고 싶으면 strip 을 사용
# strip() 은 문자열내에서 원하는 문자열 또는 공백을 제거함
# lstrip() 원하는 문자열 또는 공백을 왼쪽에서 제거
# rstrip() 원하는 문자열 또는 공백을 오른쪽에서 제거

f = open("C:/MBC/새파일.txt","r")
lines = f.readlines()
for line in lines :
    line = line.strip() #줄 끝에 줄 바꿈 \n을 제거
    print(line)
f.close()
#f. close만 쓰면 "함수객체" 차제를 출력했다라는 밑에 <function TextIOWrapper.close()> 뜸
```

```python
# read()를 이용하여 파일 문자열 읽기
# 파일의 내용 전체를 문자열로 반환
f = open("c:/MBC/새파일.txt","r")
data = f.read() # 줄별로 읽지않고 통으로가져와서 통으로 읽어서 , 한줄한줄이 아니기 때문에 \n이 먹지않고 자체처리 후 출력
print(data)
f.close()
# 전체 읽어 출력 = for문 사용 하지 X
```

```python
#for문을 이용해 줄단위로 가져오기
f = open("c:/MBC/새파일.txt","r")
# 파일 객체는 기본적으로 for문과 같이 함께 사용하여 줄단위로 읽기도 가능
for line in f :
    print(line)
f.close()
```

```python
# 파일에 새로운 내용을 추가 append -> a
# W로 파일을 열면 이미 존재하는 파일이면 그 내용이 사라지고 덮어쓰기
# a모드는 원래 있던 값을 유지하면서 새로운 값을 맨 아래에 추가

f = open("C:/MBC/새파일.txt","a") #append (추가)
f.write("======================")
for i in range (11,20):
    data = "%d번째 줄입니다\n"  %i
    f.write(data)
f.close()
```
